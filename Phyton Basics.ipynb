### Loop

```python
example=[1,5,6,'x']
n=len(example)

for i in range(n):
    print(example[i])
```

```python
for i in range(0, 8):
    print(i)
```

```python
dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while(year != 1973):
    year = dates[i]
    i = i + 1
    print(year)
```
